SVM Implementation

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

# Create Dataset

```python
from sklearn.datasets import make_classification
```

```python
# Lets create synthetic data points
X,y=make_classification(n_samples=1000,n_features=2,n_classes=2,n_clusters_per_class=2,n_redundant=0)
```

```python
X
```

```python
y
```

```python
pd.DataFrame(X)[0]
```

```python
sns.scatterplot(x=pd.DataFrame(X)[0],y=pd.DataFrame(X)[1],hue=y)
```

# Model Accuracy wrt different variant of  kernel

```python
#  Train Test Split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

## kernel Linear

```python
# Model Fitting
from sklearn.svm import SVC
svc=SVC(kernel='linear')
svc.fit(X_train, y_train)
```

```python
# Prediction
y_pred = svc.predict(X_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
print(classification_report(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
```

## kernel rbf

```python
# Model Fitting
from sklearn.svm import SVC
svc_rbf=SVC(kernel='rbf')
svc_rbf.fit(X_train, y_train)

# Prediction
y_pred_rbf = svc_rbf.predict(X_test)

from sklearn.metrics import classification_report, confusion_matrix
print(classification_report(y_test, y_pred_rbf))
print(confusion_matrix(y_test, y_pred_rbf))
```

## kernel polynomial

```python
# Model Fitting
from sklearn.svm import SVC
svc_poly=SVC(kernel='poly')
svc_poly.fit(X_train, y_train)

# Prediction
y_pred_poly = svc_poly.predict(X_test)

from sklearn.metrics import classification_report, confusion_matrix
print(classification_report(y_test, y_pred_poly))
print(confusion_matrix(y_test, y_pred_poly))
```

## kernel sigmoid

```python
# Model Fitting
from sklearn.svm import SVC
svc_sig=SVC(kernel='sigmoid')
svc_sig.fit(X_train, y_train)

# Prediction
y_pred_sig = svc_sig.predict(X_test)

from sklearn.metrics import classification_report, confusion_matrix
print(classification_report(y_test, y_pred_sig))
print(confusion_matrix(y_test, y_pred_sig))
```

# Hyperparameter Tuning

* We can play with different value of parameters to achieve more accuracy.

```python
from sklearn.model_selection import GridSearchCV
# Defining parameter range
param_grid= {'C':[0.1, 1, 10, 100, 1000],
             'gamma' : [1, 0.1, 0.01, 0.001, 0.0001],
             'kernel' : ['rbf']}
```

```python
grid = GridSearchCV(SVC(), param_grid = param_grid, cv=5, verbose=3, refit= True)
```

```python
grid.fit(X_train, y_train)
```

```python
grid.best_params_
```

```python
# Prediction
y_pred_grid= grid.predict(X_test)

from sklearn.metrics import classification_report, confusion_matrix
print(classification_report(y_test, y_pred_grid))
print(confusion_matrix(y_test, y_pred_grid))
```

In this dataset, highest accuracy is achieved by rbf kernel. It is notices that Hyperparameter tuning improved the accuracy to 95%.
